```python
import environment
import importlib

importlib.reload(environment)

from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_tavily import TavilyExtract
from langchain.agents import create_agent
```

```python
# Initialize the llm
llm = ChatGoogleGenerativeAI(
  model="gemini-3.5-flash",
  api_key=environment.GEMINI_API_KEY,
  temperature=1.0,
  max_tokens=None,
  timeout=None,
  max_retries=2,
)

# Set up the web extraction tool
tavily_extract_tool = TavilyExtract(
  extract_depth="basic",
  include_images=False,
  tavily_api_key=environment.TAVILY_API_KEY,
  # format="text",
)

system_prompt = """
You are an AI assistant, highly skilled in X (formerly Twitter) social media management.
You are able to generate X posts and threads that appear in relevant feeds and search results on the X platform, explaining and summarizing a blog post you have extracted from my technical blog website.
The X posts and threads you write are optimized for the X platform, use appropriate emojis and relevant hashtags, include a link to the extracted blog post, are written in first person from my perspective and adhere to the 240 character limit per post on the X platform.
Remember to include the book title and author I am currently studying.
"""

url = "['https://grace-sampao.github.io/posts/building-llm-powered-applications-chapter-02/']"

# Build a smart LangChain agent & bind the web extraction tool
agent = create_agent(
  model=llm,
  tools=[tavily_extract_tool],
  system_prompt=system_prompt,
)
```

```python
# Invocation
twitter_post = agent.invoke(
  {"messages": url}
)
```
