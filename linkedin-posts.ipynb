```python
import environment
import importlib

importlib.reload(environment)

from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_tavily import TavilyExtract
from langchain.agents import create_agent
```

```python
# Initialize the llm
llm = ChatGoogleGenerativeAI(
  model="gemini-3.5-flash",
  api_key=environment.GEMINI_API_KEY,
  temperature=1.0,
  max_tokens=None,
  timeout=None,
  max_retries=2,
)

# Set up the web extraction tool
tavily_extract_tool = TavilyExtract(
  extract_depth="basic",
  include_images=False,
  tavily_api_key=environment.TAVILY_API_KEY,
  # format="text",
)

system_prompt = """
You are a skilled LinkedIn social media manager.
You write a LinkedIn post, that appears in relevant feeds and search results, explaining and summarizing a blog post you have extracted from my technical blog website.
The LinkedIn post you write is optimized for the LinkedIn platform, uses appropriate emojis, includes a link to the extracted blog post and is written in first person as if I was the one who wrote it.
Remember to use relevant hashtags.
"""

url = "['https://grace-sampao.github.io/posts/building-llm-powered-applications-chapter-02/']"

# Build a smart LangChain agent & bind the web extraction tool
agent = create_agent(
  model=llm,
  tools=[tavily_extract_tool],
  system_prompt=system_prompt,
)
```

```python
# Invocation
linkedin_post = agent.invoke(
  {"messages": url}
)
```
