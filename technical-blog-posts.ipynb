```python
import environment
import time

from langchain_google_genai import ChatGoogleGenerativeAI
from google import genai
from langchain.messages import HumanMessage
```

```python
# Instantiation
model = ChatGoogleGenerativeAI(
  model="gemini-3.5-flash",
  api_key=environment.GEMINI_API_KEY,
  temperature=1.0,
  max_tokens=None,
  timeout=None,
  max_retries=2,
)

client = genai.Client(api_key=environment.GEMINI_API_KEY)
```

```python
# Upload file to Google's servers
study_notes = client.files.upload(
  file="data/Building LLM Powered Apps - Chapter 02.pdf"
)

while study_notes.state.name == "PROCESSING":
  time.sleep(2)
  study_notes = client.files.get(name=study_notes.name)
```

```python
system_message = """
You are a helpful technical writing assistant that synthesizes, edits and formats study notes in pdf format into an article in markdown format for a technical blog that uses the Jekyll Chirpy theme.
Synthesize, edit and format the user study notes in the document and output the result in markdown format.
Use approprite emojis alongside each title in the article.
"""
```

```python
# Reference by file_id in FileContentBlock
message = HumanMessage(
  content=[
    {"type": "text", "text": system_message},
    {
      "type": "file",
      "file_id": study_notes.uri,
      "mime_type": "application/pdf",
		},
	]
)
```

```python
# Invocation
response = model.invoke([message])
```

```python
print(response.content[0]["text"])
```
